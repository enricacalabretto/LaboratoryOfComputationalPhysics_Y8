1\. Find the row, column and overall means for the following matrix:

```python
m = np.arange(12).reshape((3,4))
```


````python
import numpy as np 

m = np.arange(12).reshape((3,4))
print(m)
#row
for i in m:
    mean_row=np.mean(i)
    print(f'mean for the row {i}:',mean_row)

for i in range(m.shape[1]): #shape ti da il numero di colonne 
    column= m[:,i]
    mean_column=np.mean(column)
    print(f'mean for the column{column}:', mean_column)
#overall
mean_overall= np.mean(m) 
print(f'mean of overall matrix:',mean_overall) 
````

2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
#outer in numpy
import numpy as np 

u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

outer_numpy= np.outer(u, v) 
print(f'outer with numpy:','\n', outer_numpy)

outer_listcom=np.array([[i * j for j in v] for i in u ])
print( f'outer product with list comprension:', '\n', outer_listcom)

uT= u.reshape(4,1)
print(f'broadcasting product outer:', '\n', uT * v)
````

  * Reverse the array using slice notation

````python
import numpy as np 
import matplotlib.pyplot as plt

A= np.array(np.linspace(0, 2* np.pi, 100))
print( 'matrice originale:', '\n', A)

B_10th= A[::10]
print('extract of the 10th element:', '\n', B_10th)

C_invert=A[::-1]
print(' reverse of the array', '\n', C_invert)


diff=np.abs(np.cos(A) -np.sin(A))
mask= diff<0.1
D_sincos= A[mask]
print( 'Extract of elements where the absolute difference between the sine and cosin functions evaluated at the element that is less than 0.1', '\n', D_sincos)

#make a plot of sin and cos 

y_sin= np.sin(A)
y_cos=np.cos(A)

plt.plot(A, y_sin, '-', label='sin(x)')
plt.plot(A, y_cos, '-', label='cos(x)')
plt.scatter(A[mask], y_sin[mask], color='red', label= '|sin-cos| <0.1')

plt.legend()
plt.xlabel('x')
plt.ylabel('value')
plt.show()
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import numpy as np

#ultimi due punti non capisco 
import timeit

def sieve(N):
    mask= np.ones( N+1, dtype= bool)
    mask[0:2]= False 
    for i in range( 2, int(np.sqrt(N))+1):
        if mask[i]:
            mask[i*i:N+1:i]=False 
    primes=np.arange(N+1)[mask]
    return primes
    
primes_100= sieve(100)
print('prime numbers between 0 and 99:\n', primes_100)

time_taken = timeit.timeit("sieve(10000)", setup="from __main__ import sieve", number=10)
print(f"\nTempo medio per N=10.000 (10 run): {time_taken:.6f} secondi")
````
